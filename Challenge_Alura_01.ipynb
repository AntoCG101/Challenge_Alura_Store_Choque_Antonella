#1. Análisis de facturación



### **Ingreso total por cada tienda**

#### En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
df_tienda1 = tienda['Precio'].sum()
df_tienda2 = tienda2['Precio'].sum()
df_tienda3 = tienda3['Precio'].sum()
df_tienda4 = tienda4['Precio'].sum()

print(f'El ingreso total de la primera tienda es de: ${df_tienda1}')
print(f'El ingreso total de la segunda tienda es de: ${df_tienda2}')
print(f'El ingreso total de la tercera tienda es de: ${df_tienda3}')
print(f'El ingreso total de la cuarta tienda es de: ${df_tienda4}')
```

```python
import matplotlib.pyplot as plt

nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [df_tienda1, df_tienda2, df_tienda3, df_tienda4]

plt.figure(figsize=(10, 6))

barras = plt.bar(nombres, ingresos, color='#2E8B57', edgecolor='black')

plt.title('Ingresos Totales por Tienda (Facturación)', fontsize=16)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Ingresos (en Millones de $)', fontsize=12)

for barra in barras:
    altura = barra.get_height()
    texto_etiqueta = f"${altura/1000000:.1f}M"

    plt.text(barra.get_x() + barra.get_width() / 2, altura,
             texto_etiqueta,
             ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.ticklabel_format(style='plain', axis='y')

plt.show()
```

# 3. Calificación promedio de la tienda


#### Calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
lista_tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda_1', 'Tienda_2', 'Tienda_3', 'Tienda_4']

for t, nombre in zip(lista_tiendas, nombres_tiendas):
    promedio_tienda = t['Calificación'].mean();
    promedio_redondeado = round(promedio_tienda, 3)
    print(f"Calificación promedio en {nombre}: {round(promedio_redondeado, 3)}")
```

# 4. Productos más y menos vendidos

#### En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
lista_tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda_1', 'Tienda_2', 'Tienda_3', 'Tienda_4']

for i, nombre in zip(lista_tiendas, nombres_tiendas):
  tiendas = i['Producto'].value_counts().reset_index()

  mas_vendido_nombre = tiendas.iloc[0, 0]
  mas_vendido_cantidad = tiendas.iloc[0, 1]

  menos_vendido_nombre = tiendas.iloc[-1, 0]
  menos_vendido_cantidad = tiendas.iloc[-1, 1]

  print(f'En {nombre}:')
  print(f'Los productos más vendidos son: {mas_vendido_nombre} con una cantidad de {mas_vendido_cantidad} unidades')
  print(f'Los productos menos vendidos son: {menos_vendido_nombre} con una cantidad de {menos_vendido_cantidad} unidades')
  print(" " * 50)
```

#### 4.1 Gráfico de los productos más vendidos en cada tienda

```python
import matplotlib.pyplot as plt
import numpy as np

cantidades_mas = []
nombres_mas = []
cantidades_menos = []
nombres_menos = []

for i in lista_tiendas:
    tiendas = i['Producto'].value_counts().reset_index()

    mas_vendido_nombre = tiendas.iloc[0, 0]
    mas_vendido_cantidad = tiendas.iloc[0, 1]
    nombres_mas.append(mas_vendido_nombre)
    cantidades_mas.append(mas_vendido_cantidad)

    menos_vendido_nombre = tiendas.iloc[-1, 0]
    menos_vendido_cantidad = tiendas.iloc[-1, 1]
    nombres_menos.append(menos_vendido_nombre)
    cantidades_menos.append(menos_vendido_cantidad)

x = np.arange(len(nombres_tiendas))
ancho = 0.35
plt.figure(figsize=(12, 7))

rects1 = plt.bar(x - ancho/2, cantidades_mas, ancho, label='Más Vendido', color='#99FF99')
rects2 = plt.bar(x + ancho/2, cantidades_menos, ancho, label='Menos Vendido', color='#FF9999')

plt.ylabel('Cantidad de Unidades Vendidas')
plt.title('Producto Más Vendido vs. Menos Vendido por Tienda')
plt.xticks(x, nombres_tiendas)
plt.legend()

def autolabel(rects, nombres_prod):
    for rect, nombre in zip(rects, nombres_prod):
        height = rect.get_height()
        plt.annotate(f'{height}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom', fontweight='bold')
        plt.text(rect.get_x() + rect.get_width()/2, height/2, nombre,
                 ha='center', va='center', rotation=90, color='black', fontweight='bold', fontsize=9)

autolabel(rects1, nombres_mas)
autolabel(rects2, nombres_menos)

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
lista_tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda_1', 'Tienda_2', 'Tienda_3', 'Tienda_4']

for t, nombre in zip(lista_tiendas, nombres_tiendas):
    promedio_tienda = t['Costo de envío'].mean()
    promedio_redondeado = round(promedio_tienda, 2)

    print(f"El costo de envío promedio en la {nombre} es de: ${promedio_redondeado}")
```

#### 5.1 Gráfico de los promedios de los costos de envío por tienda

```python
import matplotlib.pyplot as plt

nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
lista_tiendas = [tienda, tienda2, tienda3, tienda4]
promedios_envio = []

for e in lista_tiendas:
    promedio = e['Costo de envío'].mean()
    promedios_envio.append(promedio)

plt.figure(figsize=(10, 6))
barras = plt.bar(nombres_tiendas, promedios_envio, color=['#FF9999', '#66B3FF', '#99FF99', '#FFCC99'])

for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2, altura,
             f'${round(altura, 2)}',
             ha='center', va='bottom')

plt.title('Costo Promedio de Envío por Tienda', fontsize=16)
plt.ylabel('Costo ($)', fontsize=12)
plt.xlabel('Tiendas', fontsize=12)

# Promedio general de todas las tiendas
promedio_total = sum(promedios_envio) / len(promedios_envio)
plt.axhline(y=promedio_total, color='red', linestyle='--', label=f'Promedio General: ${round(promedio_total, 2)}')
plt.legend()

plt.show()
```
